### Data Quality Dimensions

*Accuracy*: Refers to the correctness and precision of the data. Accurate data is free from errors or discrepancies.

*Completeness*: Indicates the extent to which data is complete without any missing values or gaps. Complete data contains all the required information.

*Consistency*: Refers to the absence of contradictions or discrepancies within the data. Consistent data maintains uniformity and coherence across different sources or attributes.

*Timeliness*: Represents the relevance and currency of the data. Timely data is available when needed and reflects the current state of the information.

*Validity*: Indicates the extent to which data adheres to predefined rules, constraints, or domain-specific requirements. Valid data meets the specified criteria and conforms to the expected format or structure.

*Integrity*: Refers to the overall reliability, accuracy, and trustworthiness of the data. It ensures that the data remains intact and consistent throughout its lifecycle.

###Normalization & Standarization

Normalization, standardization,are all techniques used in data preprocessing to improve the quality and suitability of data for analysis. Here's a brief explanation of each technique:

***Data Normalization:*** It is the process of scaling numerical data to a specific range, usually between 0 and 1 or -1 and 1. Normalization helps in eliminating the scale differences between variables, making them comparable and preventing any variable from dominating the analysis due to its larger magnitude.

***Data Standardization:*** Standardization transforms numerical data to have a mean of 0 and a standard deviation of 1. It makes the data follow a standard normal distribution, which is useful for certain statistical techniques and machine learning algorithms that assume normality.

```python
from sklearn.preprocessing import MinMaxScaler

import numpy as np

# Sample dataset
data = np.array([[10], [5], [3], [8]])

# Create an instance of the MinMaxScaler
scaler = MinMaxScaler()

# Normalize the data
normalized_data = scaler.fit_transform(data)

print(normalized_data)
```

```python
# Calculate mean and std of the original data
mean_original = np.mean(data, axis=0)
std_original = np.std(data, axis=0)

print("Original Data:")
print("Mean:", mean_original)
print("Standard Deviation:", std_original)
```

```python
print("\nNormalized Data:")
print("Min Value : ", np.min(normalized_data, axis=0))
print("Max Value : ", np.max(normalized_data, axis=0))
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

# Sample dataset
data = np.array([[10], [5], [3], [8]])

# Create an instance of the StandardScaler
scaler = StandardScaler()

# Standardize the data
standardized_data = scaler.fit_transform(data)

print(standardized_data)
```

```python
print("\nStandardized Data:")
print("Mean:", np.mean(standardized_data, axis=0))
print("Standard Deviation:", np.std(standardized_data, axis=0))
```
